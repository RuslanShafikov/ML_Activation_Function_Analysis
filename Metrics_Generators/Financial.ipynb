```python
import pandas as pd

day_cpu_price = 0.075
amount_of_calls_per_day = 1000000.00
inference_time = [1.90, 1.93, 1.97, 2.05]
models = ['silu', 'gelu', 'relu', 'mish']

month_price = []
for time in inference_time:
    total_hours = (time * amount_of_calls_per_day) / 3600
    day_price = total_hours * day_cpu_price
    month_price.append(day_price * 30)

results = pd.DataFrame({
    "name": models,
    "Device": ["CPU"] * 4,
    "Inference_time": inference_time,
    "Price_per_month": month_price
})

results.to_csv("../ML/Metrics/Financial_comparison_cpu.csv", index=False)
```
